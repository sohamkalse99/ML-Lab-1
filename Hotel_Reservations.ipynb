```python
#create a day column which can include weekday, weekend, both, none

df_conditions = [(df.no_of_weekend_nights>0) & (df.no_of_week_nights>0), (df.no_of_weekend_nights>0) & (df.no_of_week_nights==0), (df.no_of_weekend_nights==0) & (df.no_of_week_nights>0), (df.no_of_weekend_nights==0) & (df.no_of_week_nights==0)]
df_categories = ["both", "weekend", "weekday", "none"]
df['day'] = np.select(df_conditions, df_categories)
df
```

Created a column day which has values ->"both", "weekend", "weekday", "none". If a person has booked the room for both weekend and weekday then the column will be filled with "both". If the room is booked only for "weekend" then it would be labeled as "weekend". If the room is booked only for "weekday" then it would be labeled as "weekday". I am segregating this using logical "AND" and logical "OR".
